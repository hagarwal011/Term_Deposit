```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the datasets
train_df = pd.read_csv("/content/train (1).csv")
test_df = pd.read_csv("/content/test (1).csv")
# Display basic info for both datasets
print("Train Data Information:\n", train_df.info())
print("Test Data Information:\n", test_df.info())
```

```python
# Checking missing values
print("Missing Values in train data:\n", train_df.isnull().sum())
print("Missing Values in test data:\n", test_df.isnull().sum())
```

```python
# Summary statistics for both datasets
print("\ntrain data Description:\n", train_df.describe())
print("\ntest data Description:\n", test_df.describe())
```

```python
# Distribution of numerical features in train data
train_df.hist(figsize=(12, 10), bins=30)
plt.suptitle("Distribution of Numerical Features in Train Data")
plt.show()

# Distribution of numerical features in test data
test_df.hist(figsize=(12, 10), bins=30)
plt.suptitle("Distribution of Numerical Features in Test Data")
plt.show()
```

```python
# Boxplots for numerical features in train data
plt.figure(figsize=(12, 6))
sns.boxplot(data=train_df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']])
plt.xticks(rotation=45)
plt.title("Boxplot of Numerical Features in Train Data")
plt.show()

# Boxplots for numerical features in test data
plt.figure(figsize=(12, 6))
sns.boxplot(data=test_df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']])
plt.xticks(rotation=45)
plt.title("Boxplot of Numerical Features in Test Data")
plt.show()
```

```python
# Correlation Heatmap for train data
plt.figure(figsize=(10, 6))
sns.heatmap(train_df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap - Train Data")
plt.show()

# Correlation Heatmap for test data
plt.figure(figsize=(10, 6))
sns.heatmap(test_df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap - Test Data")
plt.show()
```

```python
# Categorical feature distributions for train data
categorical_cols_train = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']
for col in categorical_cols_train:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=train_df, x=col, order=train_df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col} in Train Data")
    plt.show()
```

```python
# Categorical feature distributions for test data
categorical_cols_test = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
for col in categorical_cols_test:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=test_df, x=col, order=test_df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col} in Test Data")
    plt.show()
```
